```python
from shapely.geometry import Polygon, Point
import matplotlib.pyplot as plt
```

```python
corner_barrier = Polygon([(0, 4), (0, 6), (5, 6), (5, 0), (4, 0), (4, 4), (0, 4)])
p1 = Point((1, 1))
p2 = Point((7, 7))
p3 = Point((3, 3.5))
```

```python
plt.figure()
poly_x, poly_y = corner_barrier.exterior.xy
plt.plot(poly_x, poly_y)
plt.scatter(p1.x, p1.y, label="Person 1")
plt.scatter(p2.x, p2.y, label="Person 2")
plt.scatter(p3.x, p3.y, label="Person 3")
plt.legend()
plt.show()
```
